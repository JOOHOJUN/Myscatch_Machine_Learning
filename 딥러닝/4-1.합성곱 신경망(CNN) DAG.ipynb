### 합성곱 계층

>3차원 속에서 의미를 갖는 본질적인 패턴이 숨어있을 수 있으나,   
>완전연결 계층(Affine 계층)은 형상을 무시하고 모든 입력 데이터를 동등한 뉴런으로 취급
>
>합성곱 계층은 2차원, 3차원으로 데이터를 받아 학습하므로 차원 속에서 의미를 추출할 수 있다.

##### CNN용어
   1. 특징맵(feature map)
       : 합성곱 계층의 입출력 데이터
   2. 필터(filter)
       : 2차원에선 (높이,너비), 3차원에선 (채널, 높이, 너비)로 이뤄진 가중치 집합
   3. 패딩(padding)
       : 입력 데이터 주변을 특정값(보통 0)으로 추가&채우는 기법, 출력 크기를 조절할 목적으로 사용
   4. 스트라이드(stride)
       : 필터를 적용하는 간격
