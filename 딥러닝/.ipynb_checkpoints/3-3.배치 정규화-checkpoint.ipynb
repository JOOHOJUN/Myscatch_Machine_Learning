### 배치 정규화

>학습 속도 개선 / 초깃값에 크게 의존하지 않는다. / 오버피팅을 억제한다.  
배치 정규화 계층을 활성화 함수 앞(혹은 뒤)에 추가한다.  
미니배치의 가중치 분포가 평균 0, 분산 1이 되도록 한다.
